```python
#covert the rgb to gray
img2=cv2.cvtColor(img,cv2.COLOR_BGR2BGRA)
print(cv2_imshow(img2))
```

```python
#Covert image to hsv format
new_image=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(new_image)
```

```python
#Change the size of the image shape
smaller_size=cv2.resize(img,(150,150))
cv2_imshow(smaller_size)
```

```python
#Threshold Value
ret,thresh_binary=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh_binary_inv=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh_binary_trun=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh_binary_tozero=cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh_tozero_inv=cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
#Display the different threshold image
names=['original','binary','thresh_binary','Thresh_turnic','thresh_tozer','thesh_tozer_inv']
images=img,thresh_binary,thresh_binary_inv,thresh_binary_trun,thresh_binary_tozero,thresh_tozero_inv
```

```python
#for loop for the threshold image
for i in range(6):
  plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
  plt.title(names[i])
  plt.xticks([]),plt.yticks([])
```

```python
face_cascade=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
faces=face_cascade.detectMultiScale(img,1.1,4)
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
cv2_imshow(img)
cv2.waitKey()
```
